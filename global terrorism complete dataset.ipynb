```python
## TOP 10 COUNTRRIES IMPACTED BY TERRORISM:
```

```python
plt.figure(figsize= (12,8))
ei = sns.barplot(x=y.index,y=y.values,palette='crest')
for i in ei.containers:
  plt.bar_label(i,color='white')
plt.title('Top 10 countries Impacted by terrorism',color='red')
ei.axes.set_facecolor('red')
plt.gcf().set_facecolor('white')
plt.show()
```

```python
region_by_kills = df.groupby("region_txt")[["kills","nwound"]].sum().reset_index()
region_by_kills["kills"] = region_by_kills["kills"].astype("int")
region_by_kills["nwound"] = region_by_kills["nwound"].astype("int")
```

```python
region_by_kills
```

```python
plt.figure(figsize=(12,8))
ri = sns.barplot(x='region_txt',y='kills',data=region_by_kills)
for i in ri.containers:
    plt.bar_label(i)
plt.xticks(rotation=90) 
plt.title('region by kills',color='red')
plt.show()
```

```python
plt.figure(figsize=(12,8))
bii = sns.barplot(x='region_txt',y='nwound',data=region_by_kills)
for i in bii.containers:
    plt.bar_label(i)
plt.xticks(rotation=90)
plt.title('region by nwound',color='red')
plt.show()
```
