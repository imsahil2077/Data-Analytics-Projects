```python
df["date_added"] = df["date_added"].str.strip()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
df.dtypes
```

```python
df["month"] = df["date_added"].dt.month_name()
df["year"] = df["date_added"].dt.year
```

```python
df
```

```python
df.drop("show_id",axis=1,inplace=True)
```

```python
df.head()
```

```python
## rename the "listed_in" to "genre"
df.rename(columns={"listed_in":"Genre"},inplace=True)
```

```python
### genre extract
```

```python
df["Genre"] = df["Genre"].str.split(",").str[0]
```

```python
df
```

```python
df.isnull().sum()
```

```python
columnss = ["director","cast","country","date_added","rating","duration","month","year"]

for i in columnss:
    df[i] = df[i].fillna("unavailable")
```

<h1>EDA (EXPLORATORY DATA ANALYSIS)<h1/>

how many series and movies in netflix

```python
hu = sns.countplot(x="type",data=df,hue="type")
for i in hu.containers:
    plt.bar_label(i)
plt.title("movies v/s shows")
plt.show()
```

```python
df["year"].value_counts()
```

```python
plt.figure(figsize=(12,8))
df["year"].value_counts().plot(kind="bar",color="green")
```

```python
highest_months = pd.DataFrame(df["month"].value_counts())
```

```python
highest_months
```

```python
plt.figure(figsize=(12,8))
hi=sns.barplot(x="month",y="count",data=highest_months,hue="month")
for i in hi.containers:
    plt.bar_label(i,color="red")
plt.gcf().set_facecolor("black")
plt.xticks(color="white")
plt.yticks(color="white")
ko = hi.axes
ko.set_facecolor("lightblue")
plt.show()
```

```python
df["country"].value_counts().head(10)
```

```python
plt.figure(figsize=(12,8))
gi=sns.countplot(x="country",data=df,order=df["country"].value_counts().index[:10],hue="type")
for i in gi.containers:
    plt.bar_label(i)
plt.gcf().set_facecolor("lightgreen")
gi.axes.set_facecolor("lightblue")
plt.show()
```

```python
lio=df["rating"].value_counts().head(10)
```

```python
plt.figure(figsize=(12,8))
plt.pie(x=lio.values,labels=lio.index,autopct="%.1f")

center_circle = plt.Circle((0,0),radius=0.50,fc="white")
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.title("Top 10 Ratings")
plt.show()
```
