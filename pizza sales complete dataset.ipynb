```python
df['total_price'] = df['total_price'] = pd.to_numeric(df['total_price'])
```

```python
df['sales'] = df['quantity'] * df['total_price']
```

```python
df.head()
```

```python
sum_of_sales = df.groupby('month')['sales'].sum().reset_index()
```

```python
sum_of_sales
```

```python
## sum of sales by month:
```

```python
plt.figure(figsize=(12,8))
m = sns.barplot(x='month',y='sales',data=sum_of_sales,hue='month')
for i in m.containers:
  plt.bar_label(i)
plt.title('sum of sales by month',color='orange',backgroundcolor='black')

plt.show()
```

```python
## sum of sales by day:
```

```python
sum_of_sales_day = df.groupby('day')['sales'].sum().reset_index()
```

```python
sum_of_sales_day
```

```python
plt.figure(figsize=(12,8))
g= sns.barplot(x='day',y='sales',data=sum_of_sales_day,hue='day')
for i in g.containers:
  plt.bar_label(i)
plt.title('sum of sales by day',color='orange',backgroundcolor='black')
plt.show()
```
