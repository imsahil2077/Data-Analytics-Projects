```python
## cleaning the data:
```

```python
df['Launched Price (Pakistan)'] = df['Launched Price (Pakistan)'].str.replace('PKR','').str.replace(',','')
```

```python
df['Launched Price (India)'] = df['Launched Price (India)'].str.replace('INR','').str.replace(',','')
```

```python
df['Launched Price (China)'] = df['Launched Price (China)'].str.replace('CNY','').str.replace(',','').str.replace('¥','')
```

```python
df['Launched Price (USA)'] = df['Launched Price (USA)'].str.replace('USD','').str.replace(',','').str.replace('.','')
```

```python
df['Launched Price (Dubai)'] = df['Launched Price (Dubai)'].str.replace('AED','').str.replace(',','')
```

```python
## checking the null values:
```

```python
df.isnull().sum()
```

```python
## Converting it in int
```

```python
df['Launched Price (Pakistan)'] = pd.to_numeric(df['Launched Price (Pakistan)'], errors='coerce').fillna(0).astype(int)

```

```python
df['Launched Price (India)'] = pd.to_numeric(df['Launched Price (India)'],errors='coerce').fillna(0).astype(int)
```

```python
df['Launched Price (China)'] = pd.to_numeric(df['Launched Price (China)'],errors='coerce').fillna(0).astype(int)
```

```python
df['Launched Price (USA)'] = pd.to_numeric(df['Launched Price (USA)'],errors='coerce').fillna(0).astype(int)
```

```python
df['Launched Price (Dubai)']=df['Launched Price (Dubai)'].fillna(0).astype(int)
```

```python
df['Company Name'].value_counts()
```

```python
plt.figure(figsize=(12,8))
a = sns.countplot(x='Company Name',data=df,hue='Company Name')
for i in a.containers:
  plt.bar_label(i)
a.axes.set_facecolor('lightpink')
plt.xticks(rotation=90)
plt.gcf().set_facecolor('lightblue')
plt.title('Count of Mobiles',color='red')
plt.show()
```

```python
## replacing the g from mobile weight:
```

```python
df['Mobile Weight'] = df['Mobile Weight'].str.replace('g','')
```

```python
df['Mobile Weight'] = pd.to_numeric(df['Mobile Weight'],errors='coerce').fillna(0).astype(int)
```

```python
avg_mobile_weight_company = df.groupby('Company Name')['Mobile Weight'].mean().reset_index()
```

```python
avg_mobile_weight_company['Mobile Weight'] = avg_mobile_weight_company['Mobile Weight'].round(0)
```

```python
avg_mobile_weight_company
```

```python
## Average of mobile weight by mobiles:
```

```python
plt.figure(figsize=(12,8))
c = sns.barplot(x='Company Name',y='Mobile Weight',data=avg_mobile_weight_company,palette='crest')
for i in c.containers:
  plt.bar_label(i)
plt.title('Average of mobile weight by mobiles',color='blue')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Battery Capacity'] = df['Battery Capacity'].str.replace('mAh','')
```

```python
df['Battery Capacity'] = pd.to_numeric(df['Battery Capacity'],errors='coerce')
```

```python
df['Battery Capacity'].isnull().sum()
```

```python
df['Battery Capacity'] =df['Battery Capacity'].fillna(df['Battery Capacity'].mean())
```

```python
## Average of Battery capacity by Mobiles:
```

```python
avg_battery_capacity_mobiles = df.groupby('Company Name')['Battery Capacity'].mean().reset_index()
avg_battery_capacity_mobiles['Battery Capacity'] = avg_battery_capacity_mobiles['Battery Capacity'].round(0).astype(int)
```

```python


avg_battery_capacity_mobiles
```

```python
plt.figure(figsize=(12,8))
di = sns.barplot(x='Company Name',y='Battery Capacity',data=avg_battery_capacity_mobiles,palette='viridis')
for i in di.containers:
  plt.bar_label(i)
plt.title('Avg of Battery Capacity by Mobiles',color='red')
plt.xticks(rotation=90)
plt.show()
```
