```python
## remove unneccesery columns:
```

```python
a = ['lat','lng','e']
```

```python
df.drop(a,axis=1,inplace=True)
```

```python
df
```

```python
## checking null values:
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
## filling null values in zip by missing:
df['zip'] = df['zip'].fillna('missing')
```

```python
## drop the null values in twp:
```

```python
df.dropna(inplace=True)
```

```python
## rename the twp into Town:
```

```python
df.rename(columns={'twp' : 'Town'},inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
## changing timestamp into datetime

df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df.dtypes
```

```python
## extract the year from timestamp:


df['Year'] = df['timeStamp'].dt.year
```

```python
## extract month from timestamp:
```

```python
df['Month'] = df['timeStamp'].dt.month_name()
```

```python
## extract days name from timestamp:
```

```python
df['Day'] = df['timeStamp'].dt.day_name()
```

```python
df
```

```python
## extract reason from title:
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split()[0])
```

```python
df
```

```python

## Q1 what are to 5 township for 911 calls?
```

```python
plt.figure(figsize=(12,8))
ai = sns.countplot(x='Town',data=df,hue='Town',order=df['Town'].value_counts().index[:5])
for i in ai.containers:
  plt.bar_label(i)
plt.xticks(rotation=90)
ai.axes.set_facecolor ('lightblue')
plt.gcf().set_facecolor('lightgreen')
plt.title('Top 5 township for 911 calls',color='blue')
plt.show()
```

```python
## Q2 what are the top 5 zip codes for 911 calls?
```

```python
si = df['zip'].value_counts().nlargest(5)
```

```python
si
```

```python
plt.figure(figsize=(12,8))
ti = sns.barplot(x=si.index,y=si.values,hue=si.index)
for i in ti.containers:
  plt.bar_label(i)
ti.axes.set_facecolor('lightgreen')
plt.gcf().set_facecolor('pink')
plt.title('Top 5 zip codes for 911 call',color='red')
plt.show()
```
