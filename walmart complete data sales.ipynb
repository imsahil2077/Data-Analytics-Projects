```python
## COUNT OF CITY:
```

```python
ai = df['City'].value_counts().nlargest(10)
```

```python
ai
```

```python
plt.figure(figsize=(15,10))
i = sns.barplot(x=ai.index,y=ai.values,hue=ai.index)
for i in i.containers:
  plt.bar_label(i)
plt.xticks(rotation=90)
plt.title('count of top 10 cities',color='darkred')
plt.show()
```

```python
## COUNT OF STATE:
```

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(12,8))
g = sns.countplot(x='State',data=df,hue='State')
for i in g.containers:
  plt.bar_label(i)
plt.title('Count of the states',color='blue')
plt.show()
```

```python
## COUNT OF CATEGORY:
```

```python
df['Category'].value_counts()
```

```python
plt.figure(figsize=(12,8))
h = sns.countplot(x='Category',data=df,order=df["Category"].value_counts().index[::-1],hue='Category')
for i in h.containers:
  plt.bar_label(i)
plt.xticks(rotation=90)
plt.title('count of categories',color='red')
plt.show()
```

```python
## sum of sales by month:
```

```python
sum_month = df.groupby('order_month')['Sales'].sum().reset_index()
```

```python
sum_month['Sales'] = sum_month['Sales'].astype(int)
```

```python
sum_month
```

```python
plt.figure(figsize=(12,8))
t = sns.barplot(x='order_month',y='Sales',data=sum_month,hue='order_month')
for i in t.containers:
  plt.bar_label(i)
plt.title('sum of sales by month',color='red')
plt.show()
```

```python
## sum of sales by year:
```

```python
sum_sales_year = df.groupby('order_year')['Sales'].sum().reset_index()
```

```python
sum_sales_year
```

```python
plt.figure(figsize=(12,8))
ti = sns.barplot(x='order_year',y='Sales',data=sum_sales_year,hue='order_year')
for i in ti.containers:
  plt.bar_label(i)
plt.title('sum of sales by year',color='red')
plt.show()
```
